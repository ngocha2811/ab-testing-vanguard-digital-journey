# Data Cleaning

Before conducting the analysis, several data preparation steps were performed to ensure data quality and consistency:

- Checked for missing values across all datasets
- Converted variables to appropriate data types (e.g., timestamps)
- Removed duplicate records
- Merged the web activity dataset with the experiment assignment and process step datasets
- Verified that each `visit_id` corresponds to a single `client_id`
- Validated that each visit belongs to only one experiment group (Control or Test)

```python
import pandas as pd
import numpy as np
```

```python
df_web_pt1 = pd.read_csv("https://raw.githubusercontent.com/ngocha2811/ab-testing-vanguard-digital-journey/refs/heads/main/data/raw/df_final_web_data_pt_1.csv")
df_web_pt2 = pd.read_csv("https://raw.githubusercontent.com/ngocha2811/ab-testing-vanguard-digital-journey/refs/heads/main/data/raw/df_final_web_data_pt_2.csv")

df_web = pd.concat([df_web_pt1, df_web_pt2], ignore_index=True)

df_demo = pd.read_csv("https://raw.githubusercontent.com/ngocha2811/ab-testing-vanguard-digital-journey/refs/heads/main/data/raw/df_final_demo.csv")
df_variation = pd.read_csv("https://raw.githubusercontent.com/ngocha2811/ab-testing-vanguard-digital-journey/refs/heads/main/data/raw/df_final_experiment_clients.csv")
```

```python
### Cleaning df_demo

df_demo
```

```python
for col in df_demo.select_dtypes(include="number").columns:
    col_median = df_demo[col].median()
    df_demo[col].fillna(col_median, inplace=True)
```

```python
df_demo["gendr"].mode()[0]

df_demo["gendr"].fillna(df_demo["gendr"].mode()[0], inplace=True)
```

```python
df_demo.isna().sum()
```

```python
### Cleaning df_variation

df_variation.dropna(inplace=True)
```

```python
df_web
```

```python
df_web = pd.merge(left = df_web, right = df_variation, on = "client_id", how="inner")
```

```python
df_web.dropna(inplace=True)
```

```python
### Duplicates

df_web.duplicated().sum()

df_web.drop_duplicates(inplace=True)
```

```python
client_per_visit = df_web.groupby("visit_id").agg({"client_id":"nunique"}).sort_values(by="client_id", ascending=False)
client_per_visit
```

```python
visits_to_drop = client_per_visit[client_per_visit["client_id"]>1].index
visits_to_drop
```

```python
df_web = df_web[~df_web["visit_id"].isin(visits_to_drop)]
```
