### Completion Rate

```python
### Completion Rate

#---> #visit completed/#total_visits

df_web["process_step_encoded"] = df_web["process_step"].map({"start":0,
                                                             "step_1":1,
                                                             "step_2":2,
                                                             "step_3":3,
                                                             "confirm":4})

df_web
```

```python
df_cr = df_web.groupby(["Variation","visit_id"])["process_step_encoded"].max().reset_index()
df_cr["completed"] = df_cr["process_step_encoded"] == 4
df_cr
```

```python
# H0: Completion and Variation are independent
# H1: Completion and Variation are dependent

#sl = 0.05

crosstab_var_c = pd.crosstab(df_cr["Variation"], df_cr["completed"])

p_value = st.chi2_contingency(crosstab_var_c)[1]
p_value
```

```python
"""We have gathered enough evidence to reject the null hypothesis, concluding that Completion and Variation are dependent"""
```

```python
crosstab_var_c
```

```python
#We could also run a PROPORTION_ZTEST

# H0: p_Confirm Control = p_Confirm Test
# H1: p_Confirm Control != p_Confirm Test

#sl = 0.05


#number of times clients have reached "confirm step"
count = np.array([15892, 21558])

#number of observations (success + failures)
nobs = np.array([(16123+15892), (15390+21558)])

proportions_ztest(count, nobs)
```

```python
count
```

```python
#H0: p_Confirm Test - p_Confirm Control >= 5%
#H1: p_Confirm Test - p_Confirm Control < 5%

#sl=0.05

count = np.array([21558, 15892])

nobs = np.array([(15390+21558), (16123+15892)])

proportions_ztest(count, nobs, value=0.05, alternative="smaller")

```

```python
print("Completion Rate for Test: ",21558/(15390+21558))
print("Completion Rate for Control: ", 15892/(16123+15892))
```

### Error Rate

Considering Visits with Errors, if the user goes from a later step to an earlier one.

Error_rate = #visits_errors/#total_visits

```python
df_web["step_diff"] = df_web.groupby("visit_id")["process_step_encoded"].diff()
```

```python
df_error = df_web.groupby(["Variation","visit_id"])["step_diff"].min().reset_index()
df_error["is_error"] = df_error["step_diff"] < 0
df_error
```

```python
# H0: Variation and is_Error are independent
# H1: Variation and is_Error are dependent

# sl=5%

crosstab = pd.crosstab(df_error["Variation"], df_error["is_error"])
crosstab
```

```python
print("The Error Ratio on Control is ", (6475/(6475+25540)))
print("The Error Ratio on Test is ", (9914/(9914+27034)))
```

```python
st.chi2_contingency(crosstab)
```

```python
"""With such a low p_value, we can reject the null hypothesis"""
```

```python
## PROPORTION Z TEST

# H0: p_Error Test = p_Error Control
# H1: p_Error Test != p_Error Control

#sl = 5%

count = np.array([9914, 6475])

nobs = np.array([(9914+27034), (6475+25540)])


proportions_ztest(count = count , nobs=nobs)
```

```python
# H0: p_Error Test >= p_Error Control
# H1: p_Error Test < p_Error Control

#sl = 5%

count = np.array([9914, 6475])

nobs = np.array([(9914+27034), (6475+25540)])


proportions_ztest(count = count , nobs=nobs, alternative="smaller")
```

### Time spent per each Step

To calculate time spent per each step, we are going to calculate the consecutive differences on date_time per visit_id
 - DataFrame is sorted by visit_id and date_time

```python
df_web["date_time"] = pd.to_datetime(df_web["date_time"])
```

```python
df_web["time_diff"] = df_web.groupby("visit_id")["date_time"].diff()
df_web["time_diff"] = df_web["time_diff"].dt.total_seconds()
```

```python
df_web
```

```python
df_web.groupby(["Variation", "process_step"])["time_diff"].mean().reset_index()
```

```python
# H0: mu step_1 Test = mu step_1 Control
# H1: mu step_1 Test != mu step_1 Control

#sl = 5%

test_s1 = df_web[(df_web["Variation"]=="Test") & (df_web["process_step"]=="step_1")]["time_diff"].dropna()
control_s1 = df_web[(df_web["Variation"]=="Control") & (df_web["process_step"]=="step_1")]["time_diff"].dropna()

st.ttest_ind(test_s1, control_s1)
```

```python
# H0: mu confirm Test = mu confirm Control
# H1: mu confirm Test != mu confirm Control

#sl = 5%

test_s1 = df_web[(df_web["Variation"]=="Test") & (df_web["process_step"]=="confirm")]["time_diff"].dropna()
control_s1 = df_web[(df_web["Variation"]=="Control") & (df_web["process_step"]=="confirm")]["time_diff"].dropna()

st.ttest_ind(test_s1, control_s1)
```
